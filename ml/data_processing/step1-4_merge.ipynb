```python
# 데이터 병합

final_json = []
for i, json_result in enumerate(amr_guide_qwen_conversation_format):
    final_json.append(json_result)

for i, json_result in enumerate(acs_alarm_qwen_conversation_format):
    final_json.append(json_result)

for i, json_result in enumerate(mcs_textbook_qwen_conversation_format):
    final_json.append(json_result)

for i, json_result in enumerate(hacs_textbook_qwen_conversation_format):
    final_json.append(json_result)

for i, json_result in enumerate(error_scenario_textbook_qwen_conversation_format):
    final_json.append(json_result)

with open("../data/final_qwen_conversation_format.json", "w", encoding="utf-8") as f:
    json.dump(final_json, f, ensure_ascii=False, indent=4)
```

```python
len(final_json)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen3-4B-Thinking-2507")
```

```python
max_length = 0
for x in final_json:
    texts = tokenizer.apply_chat_template(x['conversation'], tokenize = True, add_generation_prompt = False)
    if len(texts) > max_length:
        max_length = len(texts)

print(max_length)
```
